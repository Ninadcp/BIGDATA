```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Seteamos el directorio en elq ue vamos a trabajar
directorio = '/Users/juanimorozumi/Documents/GitHub/BIGDATA/tp2'

#Le ponemos un nombre al archivo
archivo_nuevo = 'datos_sin_duplicados_sin_columnas.csv'

#Construimos la ruta entre el archivo y el directorio
ruta = os.path.join(directorio,archivo_nuevo)

#Cargamos el archivo en un data frame 
df = pd.read_csv(ruta)

#Mostramos las primeras filas del data frame 
print(df.head())
```

```python
#Seleccionamos las variables para la matriz de correlación
variables = ['neighbourhood_group', 'latitude', 'longitude', 'room_type', 'price', 
             'minimum_nights', 'number_of_reviews', 'reviews_per_month', 
             'calculated_host_listings_count', 'availability_365']
```

```python
# Crear una subconjunto de datos con las variables seleccionadas
subconjunto_datos = df[variables]
```

```python
# Calcular la matriz de correlación
matriz_correlacion = subconjunto_datos.corr()
```

```python
# Visualizar la matriz de correlación usando seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de correlación')
plt.show()
```
