```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Transformar todos los valores nulos a 0
df_respondieron_actualizado2.fillna(0, inplace=True)

# Convertir las características categóricas en numéricas mediante codificación one-hot
df_respondieron_actualizado2 = pd.get_dummies(df_respondieron_actualizado2, drop_first=True)

# Dividir los datos
random_state = 101
y = df_respondieron_actualizado2['pobre']
X = df_respondieron_actualizado2.drop(columns=['pobre'])

# Agregar una columna de unos a X para el término de intercepción
X = np.hstack([np.ones((X.shape[0], 1)), X])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)

# Inicializar los modelos
logit_model = LogisticRegression(random_state=random_state)
lda_model = LinearDiscriminantAnalysis()
knn_model = KNeighborsClassifier(n_neighbors=3)

# Entrenar y predecir con el modelo Logit
logit_model.fit(X_train, y_train)
y_pred_logit = logit_model.predict(X_test)
y_prob_logit = logit_model.predict_proba(X_test)[:, 1]

# Entrenar y predecir con el modelo LDA
lda_model.fit(X_train, y_train)
y_pred_lda = lda_model.predict(X_test)
y_prob_lda = lda_model.predict_proba(X_test)[:, 1]

# Entrenar y predecir con el modelo KNN
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
y_prob_knn = knn_model.predict_proba(X_test)[:, 1]

# Función para imprimir los resultados
def print_metrics(y_test, y_pred, y_prob, model_name):
    cm = confusion_matrix(y_test, y_pred)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"Resultados del modelo {model_name}:")
    print(f"Matriz de confusión:\n{cm}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"AUC: {roc_auc:.4f}\n")
    
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (área = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('Tasa de Falsos Positivos')
    plt.ylabel('Tasa de Verdaderos Positivos')
    plt.title(f'Curva ROC - {model_name}')
    plt.legend(loc="lower right")
    plt.show()

# Imprimir los resultados para cada modelo
print_metrics(y_test, y_pred_logit, y_prob_logit, "Logit")
print_metrics(y_test, y_pred_lda, y_prob_lda, "Análisis de Discriminante Lineal")
print_metrics(y_test, y_pred_knn, y_prob_knn, "KNN con k = 3")
```
