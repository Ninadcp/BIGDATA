```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
import pandas as pd

# Especifica la ruta completa del archivo Excel
file_path = '/Users/juanimorozumi/Documents/GitHub/BIGDATA/tp3/datos_filtrados.xlsx'  

# Carga el archivo Excel
data = pd.read_excel(file_path)

# Muestra las primeras filas del dataframe para verificar que se cargó correctamente
print(data.head())
```

```python
# (c) Gráfico de barras de la composición por sexo
# Asumiendo que la variable de sexo se llama 'SEXO'
sex_counts = data['CH04'].value_counts()
sex_counts.plot(kind='bar')
plt.xlabel('Sexo')
plt.ylabel('Cantidad')
plt.title('Composición por sexo')
plt.show()
```

```python
# (d) Matriz de correlación
variables = ['CH04', 'CH07', 'CH08', 'NIVEL_ED', 'ESTADO', 'CAT_INAC', 'IPCF']
correlation_matrix = data[variables].corr()
```

```python
# Graficar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de correlación')
plt.show()
```

```python
# (e) Desocupados e inactivos 
# Contar el número de desocupados e inactivos
# Asumiendo que la variable de estado ocupacional se llama 'ESTADO', donde:
# 1 representa ocupados
# 2 representa desocupados
# 3 representa inactivos

num_desocupados = data[data['ESTADO'] == 2].shape[0]
num_inactivos = data[data['ESTADO'] == 3].shape[0]

print(f'Número de desocupados: {num_desocupados}')
print(f'Número de inactivos: {num_inactivos}')

# Calcular la media de IPCF según estado ocupacional
# Asumiendo que la variable de ingreso per cápita familiar se llama 'IPCF'
mean_ipcf_by_estado = data.groupby('ESTADO')['IPCF'].mean()

print(mean_ipcf_by_estado)
```
